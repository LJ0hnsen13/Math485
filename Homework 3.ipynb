```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

import dhs_util_copy
from dhs_util_copy import *

os.chdir("C:/Users/luke/OneDrive/Data Science Course")
df = pd.read_csv('dhs_service_records_synthesized_final.csv')

df = dhs_preprocessing(df)
df, service_map = add_service_label(df)
df = add_age_bin(df)

recipient = get_recipient_attribute(df)
```

```python
df_id_serv = get_id_service_matrix(df) # this gives number of times the service is used
df_id_serv.iloc[:,1:23] = (df_id_serv.iloc[:,1:23] > 0)  # this converts value into True or False
df_id_serv
```

```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.preprocessing import *
from mlxtend.frequent_patterns import association_rules
from mlxtend.frequent_patterns import fpgrowth
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import fpmax
from mlxtend.frequent_patterns import hmine
```

```python
#extract service itemsets that have support greater than 0.01 using apriori
apriori(df_id_serv.iloc[:,1:23], use_colnames=True, min_support=0.01)\
    .sort_values(by="support", ascending=False)
```

```python
#extract service itemsets that have support greater than 0.0001 using fpgrowth
freq_itemset_fpgrowth = fpgrowth(df_id_serv.iloc[:,1:23],min_support=.0001,use_colnames=True)
freq_itemset_fpgrowth.describe()
```

```python
#extract service itemsets that have support greater than 0.0001 using hmine
hmine(df_id_serv.iloc[:,1:23], min_support=0.0001,use_colnames=True)
```

```python
# Make observations of the difference between them:

# The apriori function creates itemsets algorithmically, while the fpgrowth function only returns instances of overlap. The hmine function captures the relationships
# between the different items by creating a hyperlink tree. All of the functions return different ways of viewing relationships beteween items. 
```

```python
#extract service itemsets that have support greater than 0.0001 using fpmax
fpmax(df_id_serv.iloc[:,1:23],min_support=0.0001,use_colnames=True)
#make observations between fpmax and others
# from looking at the data, it is evident that the fpmax function looks at larger frequent itemsets and does not include any subsets of the frequent item sets. 
```

```python
#extract length-1 service itemsets that have support greater than 0.01 using apriori
apriori(df_id_serv.iloc[:,1:23], use_colnames=True, min_support=0.01, max_len=1)\
    .sort_values(by="support", ascending=False)
```

```python
#extract length-2 service itemsets that have support greater than 0.005 using hmine
freq_itemsets = hmine(df_id_serv.iloc[:,1:23], min_support=0.0001,use_colnames=True, max_len=2)
two_itemsets = freq_itemsets[freq_itemsets['itemsets'].apply(lambda x: len(x) == 2)]
two_itemsets
```

```python
#extract length-5 service itemsets that have support greater than 0.0001using fpmax
frequent_itemsets = fpmax(df_id_serv.iloc[:,1:23],min_support=0.0001,use_colnames=True)
five_itemsets = frequent_itemsets[frequent_itemsets['itemsets'].apply(lambda x: len(x) == 5)]
five_itemsets
```

```python
#extract association rules that have lift greater than 0.5, using itemsets from fpgrowth with support greater than 0.05
freq_itemsets_fpgrowth = fpgrowth(df_id_serv.iloc[:, 1:23], min_support=0.05, use_colnames=True)
asso_rules = association_rules(freq_itemsets_fpgrowth, metric="lift", min_threshold=0.5)
filtered_rules = asso_rules[asso_rules['lift'] > 0.5]
print(filtered_rules)
```

```python
#extract association rules that have support greater than 0.0003, using itemsets from fpmax with support greater than 0.0001
asso_extract = fpmax(df_id_serv.iloc[:,1:23],min_support=0.0001,use_colnames=True)
asso_rules = association_rules(asso_extract, metric ="support", min_threshold=0.0003)
print (asso_rules)
```
